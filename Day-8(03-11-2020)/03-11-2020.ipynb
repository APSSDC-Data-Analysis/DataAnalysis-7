### 1. Boxplot
- Gives us a summary of numerical data through quartiles
    - Min
    - Inter quartile Range
        - Q1 (25%)
        - Mediun (50%)
        - Q3 (75%)
    - Max
- plt.boxplot()

```python
import numpy as np
```

```python
values = np.random.randint(100,10000,1000)
values
```

```python
plt.boxplot(values)
```

```python
help(plt.boxplot)
```

- **Notched Boxplot**

```python
plt.boxplot(values, notch=True)
```

```python
laptop1 = np.random.randint(0,5000,50)
laptop2 = np.random.randint(0,10000,100)
laptop3 = np.random.randint(1000,3500,100)
```

```python
plt.boxplot([laptop1,laptop2,laptop3], labels = ['Lenovo','Dell','Apple'], 
            patch_artist = True, boxprops = dict(facecolor = 'green',color = 'r'))
plt.show()
```

### 2. Histogram
- Frequency distribution of numerical data
    - plt.hist()

```python
x = [0,0,0.1,1.2,1.2,1.3,2,2,2.5,4.1,4.2,5.0,5.1]
bins = 5
plt.hist(x, bins, rwidth = 0.4)
```

```python
x2 = [10,10,10.2,10.5,12,12.1,14,14.1,14.5,14.6,15.0,15.1]
bins = [10,12,13.5,14,15.5]
plt.hist(x2, bins, rwidth = 0.4)
```

### 3.Pie Chart
- plt.pie()

```python
games = ['Ludo','Temple Run','PubG','Mini Militia']
score = [80,90,72,50]
plt.pie(score,labels = games, autopct = '%1.2f%%',
       colors = ['r','c','m','g'], explode = (0,0.1,0,0))
plt.show()
```

```python
help(plt.pie)
```

### Color Palette
- It is an interface to generate some colors in seaborn
    - sns.color_palette()

```python
sns.color_palette()
```

```python
help(sns.color_palette)
```

```python
sns.palplot(sns.color_palette())
```

- Available seaborn palette names:
    - deep, muted, bright, pastel, dark, colorblind

```python
sns.palplot(sns.color_palette("muted"))
```

```python
sns.palplot(sns.color_palette("dark"))
```

```python
sns.palplot(sns.color_palette("RdBu", n_colors=7))
```

```python
sns.palplot(sns.color_palette("Blues", n_colors=7))
```

```python
# dark colors
sns.palplot(sns.dark_palette('purple'))
```

```python
sns.palplot(sns.dark_palette('navy'))
```

```python
# light colors
sns.palplot(sns.light_palette('navy'))
```

```python
sns.palplot(sns.color_palette('RdOr'))
```

```python
sns.palplot(sns.color_palette('RdPu_r'))
```
